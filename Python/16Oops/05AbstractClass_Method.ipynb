# Abstract Class and Abstract Method:
-------------------------------------


ABC means Abstract Base Class.

It allows you to create a template class that other classes must follow.

You cannot create an object of an abstract class.


```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
        
    @abstractmethod
    def stop(self):
        print('vehicle Stop USeing Parent Class (Abstract Class) ')

    @abstractmethod
    def applyBreak(self):
        self.model = 1985
        print('Break Model In Parent Class (Abstract Class)', self.model)
```

```python
class Bike(Vehicle):
    def start(self):
        print('Bike Start!...')

    def stop(self):
        super().stop()

    def applyBreak(self):
        super().applyBreak()
        print('Bike Model : ', self.model)
    
```

```python
objectBike = Bike()

objectBike.start()

objectBike.stop()

objectBike.applyBreak()
```

```python
class Car(Vehicle):
    def start(self):
        print('Car Start!...')

    def stop(self):
        super().stop()
        print('Car Stop')

    def applyBreak(self):
        super().applyBreak()
        print('Car Model : ', self.model)
    
```

```python
objectCar = Car()

objectCar.start()

objectCar.stop()

objectCar.applyBreak()
```
