# Date Time Module
------------------

```python
import datetime as dt
```

# Date
------

```python
curruntDate = dt.date.today()

print(curruntDate)
```

```python
userDate = dt.date(2025, 10, 21)

print(userDate)
```

Diff between two dates
----------------------

```python
diffTwoDate = curruntDate - userDate

print(diffTwoDate)
```

# Date with Time
----------------

```python
currentDateTime = dt.datetime.now()

print(currentDateTime)
```

```python
userDateTime = dt.datetime(2025, 10, 21, 8, 45, 45)

print(userDateTime)
```

Date Time Format
----------------

```python
currentDateTime = dt.datetime.now()

print(currentDateTime)

print('-----------------------------------------')

format01 = currentDateTime.strftime('%A')

print(format01)

print('-----------------------------------------')

format02 = currentDateTime.strftime('%A - %B / %d / %Y')

print(format02)
```

Split Date & Time
-----------------

```python
currentDateTime = dt.datetime.now()

print(currentDateTime)

print('-----------------------------------------')
```

```python
dateValue = currentDateTime.date()

print(dateValue)
```

```python
timeValue = currentDateTime.time()

print(timeValue)
```
