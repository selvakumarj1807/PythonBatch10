# Exception Handling or Error Handling
--------------------------------------

Try & Except Block
------------------

```python
try:
    a = 10

    b = 0

    c = a / b

    print(c)

except:
    print('Give a Valid Number!...')
```

```python
try:
    a = 10

    b = 0

    c = a / b

    print(c)

except Exception as e:
    print(e)
    print('Give a Valid Number!...')
```

Try & Except & Else
--------------------

```python
try:
    age = int(input('Enter your age : '))

except Exception as e:
    print(e)

else:
    print(age)
```

Try & Except & Else & Finally
-----------------------------

```python
try:
    age = int(input('Enter your age : '))

except Exception as e:
    print(e)

else:
    print(age)

finally:
    print('Programm Run Successfully!....')
```
