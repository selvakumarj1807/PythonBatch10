IF & ELSE
---------

```python
# Example 01

age = 12

if age >= 18: 
    print(age , ' is eligible for vote')
else:
    print(age , ' is not eligible for vote')
```

```python
# Example 02

age = 69

if age >= 18 and age <= 58: 
    print(age , ' is eligible for Job')
else:
    print(age , ' is not eligible for Job')
```

IF ELIF ELSE:
-------------

```python
age  = 12

if age > 18: 
    print(age, 'is 18+')
elif age == 18: 
    print(age, 'is 18=')
elif age == 17: 
    print(age, 'is 17')
else:
    print(age, 'is 18-')
    
```

Nested IF:
----------

```python
# Example 01

m1 = 95
m2 = 90
m3 = 95

if m1 >= 35 and m2 >= 35 and m3 >= 35:
    print('Pass!..')
    total = m1 + m2 + m2
    avg = total / 3

    print('Total : ', total)
    print('Average : ', avg)

    if avg >= 90 and avg <= 100:
        print(avg, ' A+ Grade')
    elif avg >= 80 and avg <= 90:
        print(avg, ' B+ Grade')
    else:
        print(avg, 'D Grade')

else:
    print('Fail!..')

    if m1 <= 35:
        print(m1, ' So Fail!..')
        
    if m2 <= 35:
        print(m2, ' So Fail!..')

    if m3 <= 35:
        print(m3, ' So Fail!..')
```

```python
# Example 02

m1 = 24
m2 = 34
m3 = 13

if m1 >= 35 and m2 >= 35 and m3 >= 35:
    print('Pass!..')
    total = m1 + m2 + m2
    avg = total / 3

    print('Total : ', total)
    print('Average : ', avg)

    if avg >= 90 and avg <= 100:
        print(avg, ' A+ Grade')
    elif avg >= 80 and avg <= 90:
        print(avg, ' B+ Grade')
    else:
        print(avg, 'D Grade')

else:
    print('Fail!..')

    if m1 <= 35:
        print(m1, ' So Fail!..')
        
    if m2 <= 35:
        print(m2, ' So Fail!..')

    if m3 <= 35:
        print(m3, ' So Fail!..')
```

Type Casting:
-------------

```python
a = 10

print(a, type(a))

b = float(a)

print(b, type(b))
```

```python
avg = 91.678

print(avg, type(avg))

ans = int(avg)

print(ans, type(ans))
```
