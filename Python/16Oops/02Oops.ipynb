# Create the Function In Class
------------------------------

```python
class ClassA:
    studentName = 'Student Name'
    age = 'Age'

    def details():
        print(ClassA.studentName, ' - ', ClassA.age)
```

```python
# Call the Class Function using ClassName

ClassA.details()
```

```python
class ClassB:
    studentName = 'Student Name'
    age = 'Age'

    def details(self):
        print(self.studentName, ' - ', self.age)
```

```python
student01 = ClassB()
```

```python
print(student01.studentName)
print(student01.age)
```

```python
student01.details()

student01.studentName = 'Sara'
student01.age = 25

student01.details()
```

```python
student02 = ClassB()
```

```python
student02.details()

student02.studentName = 'Lara'
student02.age = 30

student02.details()
```

```python
student01.details()

student02.details()
```

# Init Method
-------------

```python
class UserDetails:
    def __init__(self):
        print('Constructor Method for value Initial')
```

```python
obj01 = UserDetails()
```

```python
class ClassB:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print('Values Assign to the variable Successfully!...')

    def details(self):
        print(self.name, ' - ', self.age)
```

```python
student02 = ClassB('Lara', 26)
```

```python
student02.details()
```
