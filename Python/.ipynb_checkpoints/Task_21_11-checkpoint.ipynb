# 01 List Scenario:
------------------

You receive a list of product prices. Remove all prices below 100 and return the remaining list, but keep the original order.

prices = [50, 120, 300, 90, 150, 80]

Expected Output: [120, 300, 150]


# 02 Tuple Scenario:
-------------------

You get a tuple containing student details (name, age, marks). Increase marks by 10 and return a new tuple.

student = ("Arun", 20, 75)

Expected Output: ("Arun", 20, 85)


# 03 Set Scenario
-----------------

Given two sets of registered and attended employees, find who did not attend.

registered = {"ram", "sam", "john", "alex"}

attended = {"ram", "alex"}

Expected Output: {"sam", "john"}


# 04 Dictionary Scenario
------------------------

You are given a dictionary of employee salaries. Increase salary by 20% only for employees earning less than 50,000.

salaries = {
    "steve": 45000,
    "john": 52000,
    "meena": 30000
}


Example:

salary = 45000

increment = salary * 0.20

new_salary = salary + increment

print(new_salary)   # 54000.0




Expected Output:
{"steve": 54000.0, "john": 52000, "meena": 36000.0}


# 05 Mixed Data Structure Scenario
----------------------------------

You receive a list of dictionaries containing product info. Extract unique product categories using a set.

products = [

    {"name": "Laptop", "cat": "Electronics"},
    
    {"name": "Shirt", "cat": "Fashion"},
    
    {"name": "Mobile", "cat": "Electronics"},
    
    {"name": "Shoes", "cat": "Fashion"},
]

Expected Output: {"Electronics", "Fashion"}
