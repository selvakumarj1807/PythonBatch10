1. While Loop

```python
i = 0 # Value Intial

while i <= 10: # 11 <= 10 Value Conditional Ckeck
    print(i)

    i += 1 # Value Increment
```

```python
# Even number

i = 0 

while i <= 10: 
    print(i)

    i = i + 2 
```

```python
# Odd number

i = 1 

while i <= 10: 
    print(i)

    i = i + 2 
```

# Break
-------

```python
i = 0

while i <= 10:
    print(i)

    if i == 5:
        print('Before Break')
        i = i + 1
        break
        print('After Break')

    i = i + 1
```

# Continue
----------

```python
i = 0

while i <= 10:
    print(i)

    if i == 5:
        print('Before continue')
        i = i + 1
        continue
        print('After continue')

    i = i + 1
```

# Pass:
-------

```python
i = 0

while i <= 10:
    print(i)

    if i == 5:
        print('Before continue')
        i = i + 1
        pass
        print('After continue')

    i = i + 1
```
