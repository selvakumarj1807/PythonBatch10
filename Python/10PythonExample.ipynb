String Function:
----------------

```python
strValues = 'hI WelCome PyTHon'

print(strValues)

print('----------------------------------')

print(strValues.upper())

print(strValues.lower())

print(strValues.capitalize())

print(strValues.title())
```

```python
strValues = 'hi welcome python'

print(strValues)

print('----------------------------------')

print(len(strValues))

print(strValues.find('o'))

print(strValues.count('o'))
```

```python
v1 = 'admin1807'

print(v1.isupper())

print(v1.islower())

print(v1.isalpha())

print(v1.isalnum())
```

Split
-----

```python
v2 = 'Sara, Lara, Thara, Ram, Sam'

print(v2)
```

```python
listFormat = v2.split(', ')

print(listFormat)
```

Split Lines
-----------

```python
para = '''Hi 
Welcome 
Python'''

print(para)
```

```python
para = ' Hi \n Welcome \n Python'

print(para)
```

```python
para = '''Indian Army unit, also known as the Red Devils, which specializes in urban warfare and tactical needs. 
It is the oldest unit of the Indian Parachute Regiment, established in 1761 and converted to a Special Forces (SF) unit in 1978.
The battalion serves as the strategic reserve for the Special Forces and is known for its quick strikes in densely populated areas. '''

print(para)

print('------------------------------------')

listFormat = para.splitlines()

print(listFormat)
```

join:
-----

```python
listFormat = ['Sara', 'Lara', 'Thara', 'Ram', 'Sam']

print(listFormat, type(listFormat))

print('----------------------------------------------')

joinFormat = ' - '.join(listFormat)

print(joinFormat, type(joinFormat))
```

Strip:
------

```python
str01 = '  admin  '

print(len(str01))

print('---------------------')

strFormat = str01.strip()

print(strFormat, len(strFormat))
```

```python
str02 = '##admin##'

print(str02)

print(str02.strip('#'))

print(str02.lstrip('#'))

print(str02.rstrip('#'))
```
