# Tuple:
--------

1. Tupe is collection of data type. Store n number of values different data type values in a single variable.
2. Declare - ( )
3. Tuple is immutable (Unchangable)
4. Store values based on Index(starting position is 0)

```python
details = (101, 'Sara', 25, True)

print(details, type(details))
```

```python
print(details[2])
```

```python
# TypeError: 'tuple' object does not support item assignment

details[2] = 'sara@gmail.com'

print(details)
```

```python
# Type Casting
#-------------

a = 10.55
b = 20

print(a)
print(b)

print('----------------------------------')

print(int(a))

print(float(b))
```

```python
details = (101, 'Sara', 25, True)

print(details, type(details))

print('-----------------------------------')

listFormat = list(details)

print(listFormat, type(listFormat))

print('-----------------------------------')

listFormat[2] = 'sara@gmail.com'

print(listFormat, type(listFormat))

print('-----------------------------------')

details = tuple(listFormat)

print(details, type(details))
```
