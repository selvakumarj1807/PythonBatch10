List:
-----
1. List is collection of data type. Store n number of values different data type values in a single variable.
2. Declare - [ ]
3. List is mutable (changable)
4. Store values based on Index(starting position id 0)

```python
'''

0	     1	    2   	3
101 	Sara	25	  TRUE
-4	    -3	   -2	   -1

'''
```

```python
details = [101, 'Sara', 25, True]

print(details, type(details))
```

```python
# Left to Right Index Position

print(details[1], type(details[1]))
```

```python
# Right to Left Index Position

print(details[-3])
```

```python
if 'Sara' in details:
    print('Yes')
else:
    print('No')
```

```python
indx = 0

for i in details:
    print(indx, ' - ', i)

    indx += 1
```

```python
'''
ToDay Task
Question 01:

[10, 20, 22, 25, 40, 44, 43, 42]

Odd:
[25, 43]

Even:
[10, 20, 22, 40, 44, 42]

------------------------------------------------

Question 02:

[10, 20, 22, 25, 40, 44, 43, 42]

Above 18:
[20, 22, 25, 40, 44, 43, 42]

Below 18:
[10]

'''
```

```python
# Question 01

listValues = [10, 20, 22, 25, 40, 44, 43, 42]

oddValues = []
evenValues = []

for i in listValues:
    if ((i % 2) == 0):
        evenValues.append(i)
    else:
        oddValues.append(i)

print(evenValues)

print(oddValues)
```
