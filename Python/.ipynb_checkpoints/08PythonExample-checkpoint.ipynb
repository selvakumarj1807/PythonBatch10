# Dictionary
------------

1. Dictionary is a collection of data type
2. Declare - {'key': 'value'}
3. Remove Douplicate values
4. It cannot store the values index based
5. mutable

```python
user = {'id': 101, 'name': 'Sara', 'age': 25}

print(user, type(user))
```

```python
user = {'id': 101, 'name': 'Sara', 'age': 25, 'name': 'Lara'}

print(user, type(user))
```

# Get the values
----------------

```python
user = {'id': 101, 'name': 'Sara', 'age': 25}

print(user)

print('----------------------------------------')

print(user['name'])

print('----------------------------------------')

print(user.get('name'))
```

```python
user = {'id': 101, 'name': 'Sara', 'age': 25}

print(user.get('state', 'TN'))

print('----------------------------------------')

print(user['state'])
```

```python
user = {'id': 101, 'name': 'Sara', 'age': 25}

print(user)

print('-------------------')

print(user.keys())

print('-------------------')

listFormat = user.values()

print(list(listFormat))

print('-------------------')

print(user.items())
```

# Using For Loop
----------------

```python
user = {'id': 101, 'name': 'Sara', 'age': 25}

print(user)

print('-------------------')

for i in user:
    print(i, ' - ', user[i])
```

# Change Value:
---------------

```python
user = {'id': 101, 'name': 'Sara', 'age': 25}

print(user)

print('-------------------')

user['name'] = 'Lara'

print(user)
```

# Add values:
-------------

```python
user = {'id': 101, 'name': 'Sara', 'age': 25}

print(user)

print('-------------------')

user.update({'gender': 'Male', 'state': 'TN'})

print(user)
```

# Remove Values:
----------------

```python
user = {'id': 101, 'name': 'Sara', 'age': 25}

print(user)

print('-------------------')

user.pop('name')

print(user)

print('-------------------')

user.clear()

print(user)
```

# Multiple Values:
------------------

```python
list01 = [101, 'Sara', 25, [90, 95, 99]]

print(list01[1])

print(list01[3])

print(list01[3][2])
```

```python
users = {
    'user01': {'id': 101, 'name': 'Sara'},
    'user02': {'id': 102, 'name': 'Lara'}
}

print(users)
```

```python
user01_id = users['user01']['id']
user01_name = users['user01']['name']

print(user01_id, ' - ', user01_name)
```

```python
user02_id = users['user02']['id']
user02_name = users['user02']['name']

print(user02_id, ' - ', user02_name)
```
