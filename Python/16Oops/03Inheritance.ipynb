# Inheritance
-------------

Single Inheritance
------------------

```python
# Create the Class

class College:
    collegeName = 'RVS'
    webSite = 'rvs.in'

class Department(College):
    departmentName = 'Mathematics'
    studentCount = 200
```

```python
# Create the Object

object01 = Department()

print(object01.departmentName)
print(object01.studentCount)

print(object01.collegeName, ' - ', object01.webSite)
```

Multiple Inheritance:
---------------------

```python
# Create the Class

class College:
    collegeName = 'RVS'
    webSite = 'rvs.in'

class Department:
    departmentName = 'Mathematics'
    studentCount = 200

class ClassA(Department, College):
    className = 'Maths - A'
    classStudentCount = 50
```

```python
# Create the Object

object01 = ClassA()

print(object01.className, '-', object01.classStudentCount)

print(object01.departmentName, '-', object01.studentCount)

print(object01.collegeName, '-', object01.webSite)
```

Multi level Inheritance
-----------------------

```python
# Create the Class

class College:
    collegeName = 'RVS'
    webSite = 'rvs.in'

class Department(College):
    departmentName = 'Mathematics'
    studentCount = 200

class ClassA(Department):
    className = 'Maths - A'
    classStudentCount = 50
```

```python
# Create the Object

object01 = ClassA()

print(object01.className, '-', object01.classStudentCount)

print(object01.departmentName, '-', object01.studentCount)

print(object01.collegeName, '-', object01.webSite)
```
