# 1 No Return Type With Out Arguments:

```python
# Create the Function

def addTwoNum():
    a = int(input('Enter value 01: '))
    b = int(input('Enter value 02: '))

    print(a,' + ', b, ' = ', a + b)
    print('')
```

```python
# Call the Function

addTwoNum()

addTwoNum()
```

# 2 No Return type with Arguments
---------------------------------

```python
# Create the Function

def addTwoNum(a, b):
    print(a,' + ', b, ' = ', a + b)
    print('')
```

```python
# Call the Function

addTwoNum(10, 2)

addTwoNum(20, 2)
```

# 3 Return type Without Arguments:
----------------------------------

```python
# Create the Function

def addTwoNum():
    a = int(input('Enter value 01: '))
    b = int(input('Enter value 02: '))
    c = a + b
    return c
```

```python
# Call the Function

print(addTwoNum())

ans01 = addTwoNum()

print('Print Statement : ', ans01)
```

# 4 Return Type With Arguments:
-------------------------------

```python
# Create the Function

def addTwoNum(a, b):
    c = a + b
    return c
```

```python
# Call the Function

print(addTwoNum(10, 2))

v1 = addTwoNum(20, 2)

print('Print Statement : ', v1)
```

# Arbitrary Argument Function:
------------------------------

```python
# Create the Function

def details(*detail):
    print(detail, type(detail))
```

```python
# Call the Function

details(101, 'Sara', 'Male')
```

# Pass List Value in Arguments:
-------------------------------

```python
# Create the Function

def details(detail):
    print(detail, type(detail))
```

```python
# Call the Function

details([101, 'Sara', 'Male'])
```

# Key Word Argument Function
----------------------------

```python
# Create the Function

def details(studId, studName, gender):
    print('studId : ', studId)
    print('studName : ', studName)
    print('gender : ', gender)
    print('')
```

```python
# Call the Function

details(101, 'Sara', 'Male')

# Call the Function

details(studId=102, gender='Female', studName='Lara')
```

# Arbitrary Keyword Argument Function
-------------------------------------

```python
# Create the Function

def details(**detail):
    print(detail, type(detail))
   
```

```python
# Call the Function

details(studId=102, gender='Female', studName='Lara')
```

# Default Parameter Function
----------------------------

```python
# Create the Function

def details(studId, studName, gender, city = 'CHE'):
    print('studId : ', studId)
    print('studName : ', studName)
    print('gender : ', gender)
    print('city : ', city)
    print('')
```

```python
# Call the Function

details(studId=102, gender='Female', studName='Lara', city = 'CBE')


details(101, 'Sara', 'Male')
```

# Lambda Function:
------------------

```python
def addTwoNum(a, b):
    c = a + b

    return c
```

```python
print(addTwoNum(10, 2))
```

```python
# Create Lambda Function

addTwoNum = lambda a, b: a + b
```

```python
# Call the Lambda Function

print(addTwoNum(10, 2))
```
