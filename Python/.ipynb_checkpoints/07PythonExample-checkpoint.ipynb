# Set:
------

1. Set is collection of data type. Store n number of values different data type values in a single variable.
2. Declare - { }
3. Set is mutable (changable)
4. Store values based on non Index
5. Remove Douplicate Values

```python
ageValues = {10, 18, 20, 16, 25, 10, 25, 10}

print(ageValues, type(ageValues))
```

```python
for i in ageValues:
    print(i)
```

# Add Value
-----------

```python
ageValues = {10, 18, 20, 16, 25, 10, 25, 10}

print(ageValues, type(ageValues))

print('-------------------------------------------')

ageValues.add(50)

print('-------------------------------------------')

print(ageValues)
```

```python
a = {1, 3, 4, 6, 8, 0, 2, 5 }

print(a)
```

# Add Two Set Variables
-----------------------

```python
set01 = {101, 'Sara', 25, 101}

set02 = {'sara@gmai.com', True, 25}

set01.update(set02)

print(set01)
```

# Delete Values Using Discard Method:
-------------------------------------

```python
set01 = {101, 'Sara', 25, 101}

print(set01)

print('-------------------------------------')

set01.discard('Sara')

print(set01)

print('-------------------------------------')

set01.discard('sara@gmail.com')

print(set01)
```

# Delete Values Remove Method:
------------------------------

```python
set01 = {101, 'Sara', 25, 101}

print(set01)

print('-------------------------------------')

set01.remove('Sara')

print(set01)

print('-------------------------------------')

set01.remove('sara@gmail.com')

print(set01)
```

# Union
-------

```python
set01 = {1, 2, 3, 4, 5}

set02 = {4, 5, 6, 7, 8}

format01 = set01.union(set02)

print(format01)
```

# Intersect:
------------

```python
set01 = {1, 2, 3, 4, 5}

set02 = {4, 5, 6, 7, 8}

format01 = set01.intersection(set02)

print(format01)
```

# Symmetric_Difference:
-----------------------

```python
set01 = {1, 2, 3, 4, 5}

set02 = {4, 5, 6, 7, 8}

format01 = set01.symmetric_difference(set02)

print(format01)
```
