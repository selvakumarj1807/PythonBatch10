# Break
-------

```python
i = 0

while i <= 10:
    print(i)

    if i == 5:
        print('Before Break')
        i = i + 1
        break
        print('After Break')

    i = i + 1
```

# Continue
----------

```python
i = 0

while i <= 10:
    print(i)

    if i == 5:
        print('Before continue')
        i = i + 1
        continue
        print('After continue')

    i = i + 1
```

# Pass:
-------

```python
i = 0

while i <= 10:
    print(i)

    if i == 5:
        print('Before continue')
        i = i + 1
        pass
        print('After continue')

    i = i + 1
```

Nested Loop:
------------


Step 01:
i = 1
j = 0
0 < 1 True - 1
j = 1
1 < 1 False

step 02:
i = 2
j = 0
0 < 2 True - 2
j = 1
1 < 2 True - 2
j = 2
2 < 2 False


```python
for i in range(1,6):
    for j in range(i):
        print(i, end= ' ')

    print('')
```

```python
for i in range(1,6):
    for j in range(1,i+1):
        print(j, end= ' ')

    print('')
```

```python
for i in range(1,6):
    for j in range(i):
        print('*', end= ' ')

    print('')
```

```python
for i in range(5, 0, -1):
    for j in range(i):
        print(i, end=' ')

    print('')
```
