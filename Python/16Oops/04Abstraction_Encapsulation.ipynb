# Abstraction & Encapsulation
-----------------------------

```python
# Create the Class

class LibraryManagement:
    def __init__(self, books):
        self.books = books

    def displayBooks(self):
        for i in self.books:
            print(i)

    def borrowBook(self, book):
        if book in self.books:
            self.books.remove(book)
            print(book, '- Get your Book now!...')

        else:
            print(book, 'Is not there')

    def recieveBook(self, book):
        self.books.append(book)
        print(book, 'Book recieved Successfully!..')
```

```python
#### msg = ' 1. Display Book \n 2. Borrow Book \n 3. Recive Book'

bookList = ['C', 'C++', 'Java']

libraryObject01 = LibraryManagement(bookList)

while True:
    print(msg)
    print('')
    
    select = int(input('Enetre Any One value (1/2/3) : '))
    
    if select == 1:
        libraryObject01.displayBooks()
        print('')

    elif select == 2:
        book = input('Enter your Borrow Book : ')
        
        libraryObject01.borrowBook(book)
        print('')

    elif select == 3:
        book = input('Enter your Recive Book : ')

        libraryObject01.recieveBook(book)
        print('')

    else:
        print('Thank you Come Again!...')
        break
```
