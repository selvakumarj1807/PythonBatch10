# Decorater Property
--------------------

Property Getter
---------------

```python
class Student:
    def __init__(self, total):
        self._total = total

    @property
    def total(self):
        return self._total
        
```

```python
studObject01 = Student(0)
```

```python
print('Stud 01 Total : ', studObject01.total)
```

# Property Setter
-----------------

```python
class Student:
    def __init__(self, total):
        self._total = total

    @property
    def total(self):
        return self._total

    @total.setter
    def total(self, t):
        print('Setter : ', t)
```

```python
studObject01 = Student(0)
```

```python
studObject01.total = 450
```

Propety Getter & Setter
-----------------------

```python
class Student:
    def __init__(self, total):
        self._total = total

    @property
    def total(self):
        return self._total

    @total.setter
    def total(self, t):
        if t > 500 or t < 0:
            print(t, 'Assign a Valid Total')
        else:
            self._total = t
```

```python
studObject01 = Student(0)
```

```python
print('Stud 01 Total : ', studObject01.total)
```

```python
studObject01.total = 450
```

```python
print('Stud 01 Total : ', studObject01.total)
```
