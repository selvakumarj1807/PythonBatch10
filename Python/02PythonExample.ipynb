# 01 Arithmetic Operators
-------------------------

+, -, *, /, %, **, //

```python
a = 7
b = 2

print('Addtion (+) : ', a + b)
print('Subtraction (-) : ', a - b)
print('Multiplication (*) : ', a * b) # 10 * 2
print('Division (/) : ', a / b) # Return Qutiont

print('Modulas (%) : ', a % b) # Return Remainder
print('Floor Divsion (//) : ', a // b)
print('Exponentiation (**) : ', a ** b) # 7 ** 2 - 7 * 7
print('Exponentiation (**) : ', a ** 3) # 7 ** 3 - 7 * 7 * 7
```

# 02 Arithmetic Assignment Operators
------------------------------------

```python
# Assignment Operators

a = 10

print(a)

# Arithmetic Assignment Operators

a += 2 # a = a + 2

print(a)
```

# 03 Relational or Comparision Operators
----------------------------------------

<, >, <=, >=, ==, !=

```python
print(10 < 2)

print(10 > 2)

print(10 == 2)

print(10 != 2)
```

# 04 Logical Operators
----------------------

and, or, not

and - True + True = True, Other wise False

or - True + False = True, False + True = True, True + True = True, False + False = False,

not - True return False, False return True

```python
print(10 < 2)

print(10 > 2)

print((10 < 2) and (10 > 2))

print((10 < 2) or (10 > 2))

print(not(10 < 2))
```

# 05 Membership Operators
-------------------------

in, not in

```python
a = ['Sara', 'Lara', 'Ram', 'Sam']

print(a)

print('-----------------------------------------------')

print('Sara' in a)

print('sara' in a)

print('Priya' not in a)
```

# 06 Identiti Operators
-----------------------

is, is not

```python
a = 10

b = 10

print(a == b)

print(id(a))

print(id(b))

print(a is b)

print(a is not b)
```
