# Class:
--------

```python
class student:
    name = 'Sara'
    age = 25
```

```python
print(student.name)
```

```python
student.name = 'Lara'

print(student.name)
```

```python
# Get Attribute Method

print(getattr(student, 'name'))
```

```python
# Set Attribute Method

setattr(student, 'gender', 'male')

print(student.gender)
```

```python
# dell Attribute

print(student.gender)

delattr(student, 'gender')

print(student.gender)
```

```python
print(student.age)

del(student.age)

print(student.age)
```

# Class Function:
-----------------

```python
class student:
    def getData():
        name = input('Enter your name: ')
        age = int(input('Enter your age'))

        print(name, age)
```
