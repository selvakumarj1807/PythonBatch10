# Read
------

```python
try:
    fileRead = open('test.txt', 'r')

    print(fileRead.read())

    fileRead.close()

except Exception as e:
    print(e)
```

# Write
-------

```python
try:
    filewrite = open('test.txt', 'w')

    filewrite.write('Python File Handling Write Methed!...')

    filewrite.close()

    print('File Updated Successfully!...')

except Exception as e:
    print(e)
```

# Append 
--------

```python
try:
    fileAppend = open('test.txt', 'a')

    fileAppend.write(' Python File Handling Append Methed!...')

    fileAppend.close()

    print('File Updated Successfully!...')

except Exception as e:
    print(e)
```

# File Read Method

```python
try:
    fileRead = open('para.txt', 'r')

    # print(fileRead.read())
    # print(fileRead.readline())
    a = fileRead.readlines()
    print(a)

    print(type(a))
    
    fileRead.close()

except Exception as e:
    print(e)
```

# File Delete Method
--------------------

```python
import os

fileName = 'sampleRemoveFile.txt'

if os.path.exists(fileName):
    print('yes file is there!...')
    os.remove(fileName)
    print(fileName, ' file Successfully Deleted!...')

else:
    print(fileName, ' file is not in your folder!...')
```
