Data Types & Variable
---------------------

0 to 9 - Integer(int)

10.5,.. - Float(float)

'', a to z, A to Z, Special Charecters(!,@,#,$,...) - String(str) using with in ''

True, False - Boolean (bool)

```python
# Data Types & Variable

a = 10

b = 10.5

c = 'Sara'

d = True

print(a)
print(b)
print(c)
print(d)
```

```python
# To Find the Data type

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

# Concatination
---------------

```python
print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
```

```python
a = 10

b = 2

print(a + b) 

print(a, ' + ', b, ' = ', a + b) # 10 + 2 = 12
```
